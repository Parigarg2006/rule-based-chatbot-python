rule-based-chatbot-python

```python
import re
import random
from datetime import datetime

class ChatBot:
    def __init__(self, name="Arya"):
        self.name = name
        self.context = {}
        self.conversation = []
        self.rules = [
            (r'hello|hi|hey|namaste', [
                "Hello! I'm {name}. How can I help you today? 😊",
                "Hey there! What can I do for you?",
                "Namaste! How can I assist you?"
            ]),
            (r'how are you|how r u', [
                "I'm doing great, thanks for asking! How about you?",
                "Feeling fantastic! Ready to help you."
            ]),
            (r'your name|who are you', [
                f"I'm {name}, your virtual assistant!",
                f"Call me {name}! I'm here to help."
            ]),
            (r'price|cost|how much', [
                "Our prices start from ₹499. Want me to show the full price list?",
                "Pricing depends on the plan. Basic: ₹499, Pro: ₹999, Enterprise: Custom."
            ]),
            (r'order|track|delivery', [
                "To track your order, please share your order ID.",
                "Orders typically arrive in 3-5 business days. Need help tracking?"
            ]),
            (r'refund|return|cancel', [
                "Our return policy allows returns within 30 days of purchase.",
                "I can help with refunds! Please share your order number."
            ]),
            (r'help|support', [
                "I can help with: orders, pricing, returns, and general queries!",
                "Sure! Tell me what you need help with."
            ]),
            (r'thank|thanks|thankyou', [
                "You're welcome! 😊 Is there anything else?",
                "Happy to help! Feel free to ask anytime."
            ]),
            (r'bye|goodbye|exit', [
                "Goodbye! Have a great day! 👋",
                "See you later! Take care 😊"
            ]),
        ]

    def respond(self, user_input):
        user_input = user_input.lower().strip()
        self.conversation.append({"role": "user", "text": user_input})
        for pattern, responses in self.rules:
            if re.search(pattern, user_input):
                response = random.choice(responses).replace("{name}", self.name)
                self.conversation.append({"role": "bot", "text": response})
                return response
        default = random.choice([
            "I'm not sure about that. Can you rephrase?",
            "Hmm, let me connect you to a human agent for this.",
            "I don't have info on that yet. Try asking about orders, pricing, or returns!"
        ])
        self.conversation.append({"role": "bot", "text": default})
        return default

bot = ChatBot("Arya")
test_inputs = ["hello", "what is your name", "how much does it cost",
               "I want to return my order", "thank you", "bye"]
print(f"\n🤖 Chatting with {bot.name}\n")
print("="*50)
for msg in test_inputs:
    print(f"You: {msg}")
    print(f"Arya: {bot.respond(msg)}")
    print()
```
